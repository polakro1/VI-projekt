```python
import pandas as pd
import numpy as np
```

## Data preparation

### Import data

#### Import from csv

```python
data = pd.read_csv(
    "Data\EURUSD_M1.csv",
    sep=";",
    names=["time", "open", "high", "low", "close", "volume"],
    usecols=["time", "open", "high", "low", "close"],
    index_col="time",
    parse_dates=[0],
    date_format="%Y%m%d %H%M%S",
)
data.index
```

#### Reset data

```python
df = data
df.sort_index()
```

### Duplicates

#### Check for duplicates

```python
df[df.index.duplicated(keep=False)].sort_index()
```

#### Delete duplicates

```python
df = df[~df.index.duplicated(keep='first')]
df[df.index.duplicated(keep=False)].sort_index()
```

### Missing timestamps

#### Check for missing timestamps

```python
# Vygeneruj ideální časovou osu s minutovým krokem
full_index = pd.date_range(start=df.index.min(), end=df.index.max(), freq='min')

# Podmínky pro odstranění neobchodních minut (víkendy a Nový rok)
is_non_trading_time = (
    # Pátek po 17:00 až neděle do 17:00
    ((full_index.weekday == 4) & (full_index.time >= pd.to_datetime("17:00").time())) |
    (full_index.weekday == 5) |
    ((full_index.weekday == 6) & (full_index.time < pd.to_datetime("17:00").time())) |
    # Nový rok: 31.12. po 17:00 až 1.1. do 17:00
    (((full_index.month == 12) & (full_index.day == 31) & (full_index.time >= pd.to_datetime("17:00").time())) |
     ((full_index.month == 1) & (full_index.day == 1) & (full_index.time < pd.to_datetime("17:00").time()))) |
    #Vánoce: 25.12. po 03:00 až 25.12. do 17:00 
    ((full_index.month == 12) & (full_index.day == 25) & (full_index.time >= pd.to_datetime("03:00").time()) & (full_index.time <= pd.to_datetime("17:00").time()))
)

# Filtrování obchodních minut
full_index = full_index[~is_non_trading_time]

# Porovnej se skutečným indexem
missing_times = full_index.difference(df.index)
missing_df = pd.DataFrame(index=missing_times).sort_index()
missing_df
```

#### Plot missing timestamps

```python
# Vytvoř hodinový graf chybějících hodnot
missing_df['hour'] = missing_df.index.hour
missing_df.loc[missing_df.index.dayofweek == 4]['hour'].value_counts().sort_index().plot(kind='bar', figsize=(12,6))
```

#### Refill missing timestamps

```python
# Přeskládej data podle nového indexu (doplní chybějící minuty)
df_full = df.reindex(full_index)
df = df_full
```

```python
# Najdi řádky s aspoň jedním NaN
mask = df.isna().any(axis=1).to_numpy()

# Vytvoř logickou změnu mezi True/False bloky
diff = np.diff(mask.astype(int))
starts = np.where(np.insert(diff, 0, 0) == 1)[0]
ends = np.where(np.append(diff, 0) == -1)[0]

# Ošetři případ, kdy maska začíná nebo končí True
if mask[0]:
    starts = np.insert(starts, 0, 0)
if mask[-1]:
    ends = np.append(ends, len(mask) - 1)

# Sestav dataframe s výsledky
gaps = pd.DataFrame({
    'start': df.index[starts],
    'end': df.index[ends],
    'length': ends - starts + 1
})
gaps
```

#### Add Last time from last real price column

```python
mask = df.isna().any(axis=1).to_numpy()

# Najdi začátky a konce mezer
diff = np.diff(mask.astype(int))
starts = np.where(np.insert(diff, 0, 0) == 1)[0]
ends = np.where(np.append(diff, 0) == -1)[0]

if mask[0]:
    starts = np.insert(starts, 0, 0)
if mask[-1]:
    ends = np.append(ends, len(mask) - 1)

# Inicializace sloupce
time_from_last_real = np.zeros(len(df))

# Pro každou mezeru zapiš čas od poslední reálné hodnoty
for start, end in zip(starts, ends):
    for i, idx in enumerate(range(start, end + 1)):
        time_from_last_real[idx] = i + 1  # i+1 minut od poslední známé hodnoty

df['time_from_last_real_price'] = time_from_last_real
df[df["time_from_last_real_price"] > 0]
```

#### Fill NaNs

```python
fill_method = pd.Series("original", index=df.index, dtype='object')

INTERPOLATION_THRESHOLD = 5
FFILL_THRESHOLD = 30

# 1. Označ, co se má interpolovat
interp_blocks = gaps[gaps['length'] <= INTERPOLATION_THRESHOLD]
for _, row in interp_blocks.iterrows():
    fill_method.loc[row['start']:row['end']] = 'interpolate'

# 2. Označ, co se má ffillnout
ffill_blocks = gaps[(gaps['length'] > INTERPOLATION_THRESHOLD) & (gaps['length'] <= FFILL_THRESHOLD)]
for _, row in ffill_blocks.iterrows():
    fill_method.loc[row['start']:row['end']] = 'ffill'

# 3. Označ, co se má smazat
drop_blocks = gaps[gaps['length'] > FFILL_THRESHOLD]
for _, row in drop_blocks.iterrows():
    fill_method.loc[row['start']:row['end']] = 'drop'

# Default = original
df['fill_method'] = fill_method

# 4. Proveď akce podle označení
df_interp = df[df['fill_method'] == 'interpolate']
df_ffill = df[df['fill_method'] == 'ffill']
df_drop = df[df['fill_method'] == 'drop']
df_original = df[df['fill_method'] == 'original']

# Aplikuj interpolaci
for col in ['open', 'high', 'low', 'close']:
    df.loc[df_interp.index, col] = df[col].interpolate(method='linear')[df_interp.index]

# Aplikuj ffill (na základě předchozí hodnoty)
for col in ['open', 'high', 'low', 'close']:
    df.loc[df_ffill.index, col] = df[col].ffill()[df_ffill.index]

# Odstraň dlouhé mezery
df = df.drop(index=df_drop.index)
```

### Extreme and invalid values

#### Check for anomalies

```python
df['anomaly_flag'] = None

# 1. Záporné nebo nulové ceny
mask_invalid_price = (df[['open', 'high', 'low', 'close']] <= 0).any(axis=1)
df.loc[mask_invalid_price, 'anomaly_flag'] = 'invalid_price'

# 2. Extrémní rozpětí mezi high a low (> 0.01 ~ 100 pipů)
df['price_range'] = df['high'] - df['low']
mask_extreme_range = (df['price_range'] > 0.01) & df['anomaly_flag'].isna()
df.loc[mask_extreme_range, 'anomaly_flag'] = 'extreme_range'

# 3. Extrémní změna mezi minutami (close diff > 0.005)
df['price_change'] = df['close'].diff().abs()
mask_extreme_jump = (df['price_change'] > 0.005) & df['anomaly_flag'].isna()
df.loc[mask_extreme_jump, 'anomaly_flag'] = 'extreme_jump'

# 4. Logická chyba v OHLC
mask_logic_error = ((df['high'] < df['open']) |
                    (df['low'] > df['open']) |
                    (df['high'] < df['low'])) & df['anomaly_flag'].isna()
df.loc[mask_logic_error, 'anomaly_flag'] = 'logic_error'

# Shrnutí
anomalies = df[df['anomaly_flag'].notna()]
anomalies
```
