### Missing timestamps

#### Check for missing timestamps

```python
# Vygeneruj ideální časovou osu s minutovým krokem
full_index = pd.date_range(start=df.index.min(), end=df.index.max(), freq="min")

weekdays = full_index.weekday
times = full_index.time
months = full_index.month
days = full_index.day

# Podmínky pro odstranění neobchodních minut (víkendy a Nový rok)
is_non_trading_time = (
    # Pátek po 17:00 až neděle do 17:00
    ((weekdays == 4) & (times >= pd.Timestamp("17:00").time()))
    | (weekdays == 5)
    | ((weekdays == 6) & (times < pd.Timestamp("17:00").time()))
    |
    # Nový rok: 31.12. po 17:00 až 1.1. do 17:00
    (
        ((months == 12) & (days == 31) & (times >= pd.Timestamp("17:00").time()))
        | ((months == 1) & (days == 1) & (times < pd.Timestamp("17:00").time()))
    )
    |
    # Vánoce: 25.12. po 03:00 až 25.12. do 17:00
    (
        (months == 12)
        & (days == 25)
        & (times >= pd.Timestamp("03:00").time())
        & (times <= pd.Timestamp("17:00").time())
    )
)

# Filtrování obchodních minut
full_index = full_index[~is_non_trading_time]

# Porovnej se skutečným indexem
missing_times = full_index.difference(df.index)
missing_df = pd.DataFrame(index=missing_times).sort_index()
missing_df
```

#### Gaps


##### Localize gaps

```python
# Najdi řádky s aspoň jedním NaN
mask = df.isna().any(axis=1).to_numpy()

# Vytvoř logickou změnu mezi True/False bloky
diff = np.diff(mask.astype(int))
starts = np.where(np.insert(diff, 0, 0) == 1)[0]
ends = np.where(np.append(diff, 0) == -1)[0]

# Ošetři případ, kdy maska začíná nebo končí True
if mask[0]:
    starts = np.insert(starts, 0, 0)
if mask[-1]:
    ends = np.append(ends, len(mask) - 1)

# Sestav dataframe s výsledky
gaps = pd.DataFrame(
    {"start": df.index[starts], "end": df.index[ends], "length": ends - starts + 1}
)
gaps
```

#### Add Last time from last real price column

```python
mask = df.isna().any(axis=1).to_numpy()

# Najdi začátky a konce mezer
diff = np.diff(mask.astype(int))
starts = np.where(np.insert(diff, 0, 0) == 1)[0]
ends = np.where(np.append(diff, 0) == -1)[0]

if mask[0]:
    starts = np.insert(starts, 0, 0)
if mask[-1]:
    ends = np.append(ends, len(mask) - 1)

# Inicializace sloupce
time_from_last_real = np.zeros(len(df))

# Pro každou mezeru zapiš čas od poslední reálné hodnoty
for start, end in zip(starts, ends):
    for i, idx in enumerate(range(start, end + 1)):
        time_from_last_real[idx] = i + 1  # i+1 minut od poslední známé hodnoty

df["time_from_last_real_price"] = time_from_last_real
df[df["time_from_last_real_price"] > 0]
```

#### Fill NaNs

```python
fill_method = pd.Series("original", index=df.index, dtype="object")

INTERPOLATION_THRESHOLD = 5
FFILL_THRESHOLD = 30

# 1. Označ, co se má interpolovat
interp_blocks = gaps[gaps["length"] <= INTERPOLATION_THRESHOLD]
for _, row in interp_blocks.iterrows():
    fill_method.loc[row["start"] : row["end"]] = "interpolate"

# 2. Označ, co se má ffillnout
ffill_blocks = gaps[
    (gaps["length"] > INTERPOLATION_THRESHOLD) & (gaps["length"] <= FFILL_THRESHOLD)
]
for _, row in ffill_blocks.iterrows():
    fill_method.loc[row["start"] : row["end"]] = "ffill"

# 3. Označ, co se má smazat
drop_blocks = gaps[gaps["length"] > FFILL_THRESHOLD]
for _, row in drop_blocks.iterrows():
    fill_method.loc[row["start"] : row["end"]] = "drop"

# Default = original
df["fill_method"] = fill_method

# 4. Proveď akce podle označení
df_interp = df[df["fill_method"] == "interpolate"]
df_ffill = df[df["fill_method"] == "ffill"]
df_drop = df[df["fill_method"] == "drop"]
df_original = df[df["fill_method"] == "original"]

# Aplikuj interpolaci
for col in ["open", "high", "low", "close"]:
    df.loc[df_interp.index, col] = df[col].interpolate(method="linear")[df_interp.index]

# Aplikuj ffill (na základě předchozí hodnoty)
for col in ["open", "high", "low", "close"]:
    df.loc[df_ffill.index, col] = df[col].ffill()[df_ffill.index]

# Odstraň dlouhé mezery
df = df.drop(index=df_drop.index)
```

### Extreme and invalid values

#### Check for anomalies

```python
df["anomaly_flag"] = None

# 1. Záporné nebo nulové ceny
mask_invalid_price = (df[["open", "high", "low", "close"]] <= 0).any(axis=1)
df.loc[mask_invalid_price, "anomaly_flag"] = "invalid_price"

# 2. Extrémní rozpětí mezi high a low (> 0.01 ~ 100 pipů)
df["price_range"] = df["high"] - df["low"]
mask_extreme_range = (df["price_range"] > 0.01) & df["anomaly_flag"].isna()
df.loc[mask_extreme_range, "anomaly_flag"] = "extreme_range"

# 3. Extrémní změna mezi minutami (close diff > 0.005)
df["price_change"] = df["close"].diff().abs()
mask_extreme_jump = (df["price_change"] > 0.005) & df["anomaly_flag"].isna()
df.loc[mask_extreme_jump, "anomaly_flag"] = "extreme_jump"

# 4. Logická chyba v OHLC
mask_logic_error = (
    (df["high"] < df["open"]) | (df["low"] > df["open"]) | (df["high"] < df["low"])
) & df["anomaly_flag"].isna()
df.loc[mask_logic_error, "anomaly_flag"] = "logic_error"

# Shrnutí
anomalies = df[df["anomaly_flag"].notna()]
anomalies
```

# Feature Engineering

## Calculate features
ema, rsi, bollinger bands, hour, day of week, macd, indikátor volatilit

```python
#EMA
df["EMA_5"] = talib.EMA(df["close"], timeperiod=5)
df["EMA_15"] = talib.EMA(df["close"], timeperiod=15)
df["EMA_60"] = talib.EMA(df["close"], timeperiod=60)

#RSI
df["RSI_14"] = talib.RSI(df["close"], timeperiod=14)

#Bollinger bands
upper, middle, lower = talib.BBANDS(df["close"], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
df["Bollinger_middle"] = middle
df["Bollinger_upper_diff"] = upper - df["close"]
df["Bollinger_lower_diff"] = df["close"] - lower
df["Bollinger_width"] = upper - lower

#MACD
macd, macdsignal, macdhist = talib.MACD(df["close"], fastperiod=12, slowperiod=26, signalperiod=9)
df["MACD"] = macd
df["MACD_signal"] = macdsignal
df["MACD_diff"] = macd - macdsignal

#ATR
df["ATR_14"] = talib.ATR(df["high"], df["low"], df["close"], timeperiod=14)

#Hour
df["hour"] = df.index.hour
df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)

#Day of week
df["dayofweek"] = df.index.dayofweek
df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

df
```

## Target

```python
future_offset = to_offset("15min")

# Vytvoř nový sloupec se zavírací cenou za 15 minut (nejbližší další)
df["future_time"] = df.index + future_offset
df["future_close"] = df["close"].reindex(df["future_time"]).values


# Inicializace target jako NaN
df["target_15m"] = np.nan

# Cíl pro klasifikaci – binární směr pohybu
valid_rows = df["future_close"].notna()
df.loc[valid_rows, "target_15m"] = (
    df.loc[valid_rows, "future_close"] > df.loc[valid_rows, "close"]
).astype(int)

columns_to_ignore = ['anomaly_flag']
columns_to_use = list(set(df.columns) - set(columns_to_ignore))

rows_with_nan = df[df["target_15m"].isna()]
rows_with_nan
```

## Drop NaNs

## Normalization

```python
# === Vstupní sloupce podle typu transformace ===

# 1. StandardScaler – běžné technické indikátory a ceny
standard_cols = [
    "open", "high", "low", "close",
    "price_range",
    "EMA_5", "EMA_15", "EMA_60",
    "MACD", "MACD_signal", "MACD_diff",
    "Bollinger_middle", "Bollinger_upper_diff", "Bollinger_lower_diff",
    "ATR_14"
]

# 2. MinMaxScaler – RSI (přirozený rozsah 0–100)
minmax_cols = ["RSI_14"]

# 3. log + StandardScaler – čas od poslední reálné ceny
log_scaled_cols = ["time_from_last_real_price"]

# 4. Cyklické proměnné – už jsou připravené jako sin/cos → použijeme StandardScaler
cyclical_cols = ["hour_sin", "hour_cos", "dow_sin", "dow_cos"]

# === Aplikace transformací ===

# Log transformace
df.loc[:, "log_time_since_real"] = np.log1p(df["time_from_last_real_price"])

# Scalers
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
log_scaler = StandardScaler()

# Fit-transform
standard_scaled = standard_scaler.fit_transform(df[standard_cols])
minmax_scaled = minmax_scaler.fit_transform(df[minmax_cols])
log_scaled = log_scaler.fit_transform(df[["log_time_since_real"]])
cyclical_data = df[cyclical_cols].values  # reuse standard_scaler

# === Přidání do df_clean s "_scaled" suffixem ===

# Helper pro přidání sloupců
def add_scaled_columns(df, scaled_array, columns, suffix="_scaled"):
    for i, col in enumerate(columns):
        df.loc[:, col + suffix] = scaled_array[:, i]

add_scaled_columns(df, standard_scaled, standard_cols)
add_scaled_columns(df, minmax_scaled, minmax_cols)
add_scaled_columns(df, log_scaled, ["log_time_since_real"])
add_scaled_columns(df, cyclical_data, cyclical_cols)

df
```

```python
X_train = train_df[feature_columns].values
y_train = train_df["target_15m"].values

X_val = val_df[feature_columns].values
y_val = val_df["target_15m"].values

X_test = test_df[feature_columns].values
y_test = test_df["target_15m"].values
```

## Creating sliding windows

```python
def create_sliding_windows(X, y, window_size):
    # Vytvoření sliding window view
    X_windows = np.lib.stride_tricks.sliding_window_view(X, window_shape=(window_size, X.shape[1]))
    
    # Odstranění nadbytečné dimenze (squeeze)
    X_windows = X_windows.squeeze(1)

    # Targety (musí začínat až za posledním oknem)
    y_windows = y[window_size:]
    
    return X_windows, y_windows
```

```python
window_size = 60  # např. 60 minut

# Trénovací
X_train_seq, y_train_seq = create_sliding_windows(X_train, y_train, window_size=60)
#Validační
X_val_seq, y_val_seq = create_sliding_windows(X_val, y_val, window_size=60)
#Trénovací
X_test_seq, y_test_seq = create_sliding_windows(X_test, y_test, window_size=60)
```

# Neural Network

```python
# Příklad: window = 60 minut, 22 vstupních rysů
time_steps = 60  
input_features = 22

model = Sequential()
# První LSTM vrstva
model.add(LSTM(64, return_sequences=True, input_shape=(time_steps, input_features)))
model.add(Dropout(0.3))

# Druhá LSTM vrstva
model.add(LSTM(32, return_sequences=False))
model.add(Dropout(0.3))

# Dense vrstvy
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))

# Výstupní vrstva
model.add(Dense(1, activation='sigmoid'))

# Kompilace modelu
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```
