```python
import pandas as pd
import numpy as np
import talib
from pandas.tseries.frequencies import to_offset
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

## Data preparation

```python
data = pd.read_csv(
        "Data\EURUSD_M1.csv",
        sep=";",
        names=["timestamp", "open", "high", "low", "close", "volume"],
        usecols=["timestamp", "open", "high", "low", "close"],
        parse_dates=[0],
        date_format="%Y%m%d %H%M%S",)
data
```

### NaNs

#### Show NaNs

```python
columns_to_ignore = ['anomaly_flag']
columns_to_use = list(set(df.columns) - set(columns_to_ignore))

rows_with_nan = df[df[columns_to_use].isna().any(axis=1)]
rows_with_nan
```

#### Fill NaNs

### Extreme and invalid values

#### Check for anomalies

```python
df["anomaly_flag"] = None

# 1. Záporné nebo nulové ceny
mask_invalid_price = (df[["open", "high", "low", "close"]] <= 0).any(axis=1)
df.loc[mask_invalid_price, "anomaly_flag"] = "invalid_price"

# 2. Extrémní rozpětí mezi high a low (> 0.01 ~ 100 pipů)
df["price_range"] = df["high"] - df["low"]
mask_extreme_range = (df["price_range"] > 0.01) & df["anomaly_flag"].isna()
df.loc[mask_extreme_range, "anomaly_flag"] = "extreme_range"

# 3. Extrémní změna mezi minutami (close diff > 0.005)
df["price_change"] = df["close"].diff().abs()
mask_extreme_jump = (df["price_change"] > 0.005) & df["anomaly_flag"].isna()
df.loc[mask_extreme_jump, "anomaly_flag"] = "extreme_jump"

# 4. Logická chyba v OHLC
mask_logic_error = (
    (df["high"] < df["open"]) | (df["low"] > df["open"]) | (df["high"] < df["low"])
) & df["anomaly_flag"].isna()
df.loc[mask_logic_error, "anomaly_flag"] = "logic_error"

# Shrnutí
anomalies = df[df["anomaly_flag"].notna()]
anomalies
```

# Feature Engineering

## Calculate features
ema, rsi, bollinger bands, hour, day of week, macd, indikátor volatilit

```python
#EMA
df["EMA_5"] = talib.EMA(df["close"], timeperiod=5)
df["EMA_15"] = talib.EMA(df["close"], timeperiod=15)
df["EMA_60"] = talib.EMA(df["close"], timeperiod=60)

#RSI
df["RSI_14"] = talib.RSI(df["close"], timeperiod=14)

#Bollinger bands
upper, middle, lower = talib.BBANDS(df["close"], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
df["Bollinger_middle"] = middle
df["Bollinger_upper_diff"] = upper - df["close"]
df["Bollinger_lower_diff"] = df["close"] - lower
df["Bollinger_width"] = upper - lower

#MACD
macd, macdsignal, macdhist = talib.MACD(df["close"], fastperiod=12, slowperiod=26, signalperiod=9)
df["MACD"] = macd
df["MACD_signal"] = macdsignal
df["MACD_diff"] = macd - macdsignal

#ATR
df["ATR_14"] = talib.ATR(df["high"], df["low"], df["close"], timeperiod=14)

#Hour
df["hour"] = df["timestamp"].dt.hour
df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)

#Day of week
df["dayofweek"] = df["timestamp"].dt.dayofweek
df["dow_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
df["dow_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

df
```

## Target

```python
# Posun zavírací ceny o 1 řádek dopředu (např. 15 minutový krok)
df["future_close"] = df["close"].shift(-1)

# Inicializace target jako NaN
df["target_15m"] = np.nan

# Cíl pro klasifikaci – binární směr pohybu
valid_rows = df["future_close"].notna()
df.loc[valid_rows, "target_15m"] = (
    df.loc[valid_rows, "future_close"] > df.loc[valid_rows, "close"]
).astype(int)

columns_to_ignore = ['anomaly_flag']
columns_to_use = list(set(df.columns) - set(columns_to_ignore))

# Řádky, kde target není definován (na konci datasetu)
rows_with_nan = df[df["target_15m"].isna()]
rows_with_nan
```

## Drop NaNs

```python
# Seznam sloupců, které mají být bez NaN (indikátory)
indicator_cols = [
    "EMA_5", "EMA_15", "EMA_60",
    "RSI_14",
    "MACD", "MACD_signal", "MACD_diff",
    "Bollinger_middle", "Bollinger_upper_diff", "Bollinger_lower_diff",
    "ATR_14",
    "target_15m"
]

count = df[indicator_cols].isna().any(axis=1).sum()

# Očistit jen podle těchto sloupců
df = df.dropna(subset=indicator_cols)

count
```

## Normalization

```python
# === Vstupní sloupce podle typu transformace ===

# 1. StandardScaler – běžné technické indikátory a ceny
standard_cols = [
    "open", "high", "low", "close",
    "price_range",
    "EMA_5", "EMA_15", "EMA_60",
    "MACD", "MACD_signal", "MACD_diff",
    "Bollinger_middle", "Bollinger_upper_diff", "Bollinger_lower_diff",
    "ATR_14"
]

# 2. MinMaxScaler – RSI (přirozený rozsah 0–100)
minmax_cols = ["RSI_14"]

# 3. log + StandardScaler – čas od poslední reálné ceny
log_scaled_cols = ["time_from_last_real_price"]

# 4. Cyklické proměnné – už jsou připravené jako sin/cos → použijeme StandardScaler
cyclical_cols = ["hour_sin", "hour_cos", "dow_sin", "dow_cos"]

# === Aplikace transformací ===

# Log transformace

# Scalers
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
log_scaler = StandardScaler()

# Fit-transform
standard_scaled = standard_scaler.fit_transform(df[standard_cols])
minmax_scaled = minmax_scaler.fit_transform(df[minmax_cols])
cyclical_data = df[cyclical_cols].values  # reuse standard_scaler

# === Přidání do df_clean s "_scaled" suffixem ===

# Helper pro přidání sloupců
def add_scaled_columns(df, scaled_array, columns, suffix="_scaled"):
    for i, col in enumerate(columns):
        df.loc[:, col + suffix] = scaled_array[:, i]

add_scaled_columns(df, standard_scaled, standard_cols)
add_scaled_columns(df, minmax_scaled, minmax_cols)
add_scaled_columns(df, cyclical_data, cyclical_cols)

# Normalizace targetu (regrese)
target_scaler = StandardScaler()
target_scaled = target_scaler.fit_transform(df[["future_close"]])
df["future_close_scaled"] = target_scaled

df
```

# Preparing dataset for NN

## Feature columns

```python
feature_columns = [col for col in df.columns if col.endswith("_scaled")]
feature_columns.remove("future_close_scaled")
```

## Splitting dataset

```python
# Urči poměry
train_frac = 0.7
val_frac = 0.15
test_frac = 0.15

# Celkový počet vzorků
n = len(df)

# Výpočty dělicích bodů
train_end = int(n * train_frac)
val_end = train_end + int(n * val_frac)

# Rozdělení
train_df = df.iloc[:train_end]
val_df = df.iloc[train_end:val_end]
test_df = df.iloc[val_end:]
```

```python
feature_columns
```

```python
X_train = train_df[feature_columns].values
y_train = train_df["target_15m"].values

X_val = val_df[feature_columns].values
y_val = val_df["target_15m"].values

X_test = test_df[feature_columns].values
y_test = test_df["target_15m"].values
```

## Creating sliding windows

```python
def create_sliding_windows(X, y, window_size):
    # Vytvoření sliding window view
    X_windows = np.lib.stride_tricks.sliding_window_view(X, window_shape=(window_size, X.shape[1]))
    
    # Odstranění nadbytečné dimenze (squeeze)
    X_windows = X_windows.squeeze(1)

    # Targety (musí začínat až za posledním oknem)
    y_windows = y[window_size - 1:]
    
    return X_windows, y_windows

window_size = 60  # např. 60 minut

# Trénovací
X_train_seq, y_train_seq = create_sliding_windows(X_train, y_train, window_size)
#Validační
X_val_seq, y_val_seq = create_sliding_windows(X_val, y_val, window_size)
#Trénovací
X_test_seq, y_test_seq = create_sliding_windows(X_test, y_test, window_size)
```

```python
print(X_train_seq.shape)
print(y_train_seq.shape)
```

# Neural Network

```python
# Příklad: window = 60 minut, 21 vstupních rysů
time_steps = 60  
input_features = 21
opt = Adam(learning_rate=0.01, decay=0)

model = Sequential()
# První LSTM vrstva
model.add(LSTM(64, return_sequences=True, input_shape=(X_test_seq.shape[1], X_test_seq.shape[2])))
model.add(Dropout(0.3))

# Druhá LSTM vrstva
model.add(LSTM(32, return_sequences=False, input_shape=(X_test_seq.shape[1], X_test_seq.shape[2]), activation= "tanh"))
model.add(Dropout(0.3))

# Dense vrstvy
model.add(Dense(16, activation='relu'))
model.add(Dropout(0.2))

# Výstupní vrstva
model.add(Dense(1, activation='sigmoid'))

# Kompilace modelu
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

## Training

```python
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train_seq, y_train_seq,
    validation_data=(X_val_seq, y_val_seq),
    epochs=50,
    batch_size=64,
    callbacks=[early_stop, reduce_lr],
    verbose = 1
)
```
